```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.graph_objects as go
from prophet import Prophet

# Fetch historical stock data for "AAPL" ticker
ticker = input("Enter ticker symbols (comma-separated): ").split(",")
ticker = [ticker.strip().upper() for ticker in ticker]
stock_data = yf.download(ticker, start="2019-01-01", end="2023-07-19")  # Replace "today" with the current date or any specific end date.

# Prepare the data for Prophet
prophet_data = stock_data.reset_index()[["Date", "Close"]]
prophet_data.rename(columns={"Date": "ds", "Close": "y"}, inplace=True)

# Create and fit the Prophet model
model = Prophet(daily_seasonality=True)
model.fit(prophet_data)

# Make future predictions using the Prophet model
future = model.make_future_dataframe(periods=365)  # 30 days forecast
forecast = model.predict(future)

# Calculate Bollinger Bands
window = 20
std_dev = 2
stock_data["MA"] = stock_data["Close"].rolling(window=window).mean()
stock_data["Upper"] = stock_data["MA"] + std_dev * stock_data["Close"].rolling(window=window).std()
stock_data["Lower"] = stock_data["MA"] - std_dev * stock_data["Close"].rolling(window=window).std()

# Identify bearish and bullish signals
bullish_signals = stock_data[stock_data["Close"] > stock_data["Upper"]]["Close"]
bearish_signals = stock_data[stock_data["Close"] < stock_data["Lower"]]["Close"]

# Create a Plotly figure
fig = go.Figure()

# Add Bollinger Bands
fig.add_trace(go.Scatter(x=stock_data.index, y=stock_data["Upper"], mode="lines", name="Upper Bollinger Band", line=dict(color="blue")))
fig.add_trace(go.Scatter(x=stock_data.index, y=stock_data["Lower"], mode="lines", name="Lower Bollinger Band", line=dict(color="blue")))

# Add Closing Price
fig.add_trace(go.Scatter(x=stock_data.index, y=stock_data["Close"], mode="lines", name="Closing Price", line=dict(color="black")))

# Add Prophet forecast
fig.add_trace(go.Scatter(x=forecast["ds"], y=forecast["yhat"], mode="lines", name="Prophet Forecast", line=dict(color="purple")))

# Add bullish signals
fig.add_trace(go.Scatter(x=bullish_signals.index, y=bullish_signals, mode="markers", name="Bullish Signal", marker=dict(symbol="triangle-up", size=10, color="green")))

# Add bearish signals
fig.add_trace(go.Scatter(x=bearish_signals.index, y=bearish_signals, mode="markers", name="Bearish Signal", marker=dict(symbol="triangle-down", size=10, color="red")))

# Add trend line with forecast points
fig.add_trace(go.Scatter(x=forecast["ds"], y=forecast["trend"], mode="lines", name="Trend Line", line=dict(color="orange", dash="dash")))

# Update the layout
fig.update_layout(title=f"{ticker} Bollinger Bands, Prophet Forecast, and Bullish/Bearish Signals",
                  xaxis_title="Date",
                  yaxis_title="Price",
                  legend=dict(x=0, y=1),
                  template="plotly_white")

# Show the chart
fig.show()
```
